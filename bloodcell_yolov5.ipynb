```python
def convert(line):
  li = line.split()
  li_sub = li
  if li[0] == 'rbc':
    li_sub[0] = '0'
  else:
    li_sub[0] = '1'
  for i in range(4,8):
    li[i] = float(li[i]) / 256
  li_sub[1] = (li[4] + li[6]) / 2
  li_sub[2] = (li[5] + li[7]) / 2
  li_sub[3] = li[6] - li[4]
  li_sub[4] = li[7] - li[5]
  s = '{} {} {} {} {}'
  s = s.format(li_sub[0],li_sub[1],li_sub[2],li_sub[3],li_sub[4])
  return s
convert('wbc 0.00 0 0 85.610268378063 2.334889148191365 187.7386231038506 120.0688448074679 0 0 0 0 0 0 0')
  


```
